# K-Nearest Neighbours

```python
import numpy as np
import matplotlib.pyplot as plt
```

```python
# creating distributions
mean_01 = np.array([0.0, 0.0])
cov_01 = np.array([[1, 0.1], [0.1, 1]])
dist_01 = np.random.multivariate_normal(mean_01, cov_01, 500)

mean_02 = np.array([5.0, 5.0])
cov_02 = np.array([[1, -0.5], [-0.5, 1]])
dist_02 = np.random.multivariate_normal(mean_02, cov_02, 500)

mean_03 = np.array([6.0, -1.0])
cov_03 = np.array([[1, -0.5], [-0.5, 1]])
dist_03 = np.random.multivariate_normal(mean_03, cov_03, 500)
```

```python
# visualising data
plt.scatter(dist_01[:, 0], dist_01[:, 1], marker = '.', label = 'Class 0')
plt.scatter(dist_02[:, 0], dist_02[:, 1], marker = '.', label = 'Class 1')
plt.scatter(dist_03[:, 0], dist_03[:, 1], marker = '.', label = 'Class 2')
plt.legend()
plt.show()
```

```python
# create data
rows = dist_01.shape[0] + dist_02.shape[0] + dist_03.shape[0]
cols = dist_01.shape[1] + 1

data = np.zeros((rows, cols))

data[:dist_01.shape[0], :-1] = dist_01
data[dist_01.shape[0]:dist_01.shape[0] + dist_02.shape[0], :-1] = dist_02
data[dist_01.shape[0] + dist_02.shape[0]: , :-1] = dist_03
data[dist_01.shape[0]:dist_01.shape[0] + dist_02.shape[0], -1] = 1.0
data[dist_01.shape[0] + dist_02.shape[0]: , -1] = 2.0
```

```python
# train test split
np.random.shuffle(data)
split_factor = 0.85
split = int(split_factor*data.shape[0])
train = data[:split, :]
test = data[split: , :]

X_train = train[:, :-1]
y_train = train[:, -1]
X_test = test[:, :-1]
y_test = test[:, -1]
```

```python
# distance function - euclidean distance function
def distance(p1, p2):
    return np.sqrt(np.sum((p1 - p2)**2))
```

```python
# k nearest neighbors
def knn(test_point, X_train, y_train, k = 5):
    d = []
    for i in range(X_train.shape[0]):
        dist = distance(test_point, X_train[i, :])
        d.append((dist, y_train[i]))
    d = sorted(d)
    sorted_d = np.array(d[:k])
    freq = np.unique(sorted_d[:, -1], return_counts = True)
    label = freq[0][freq[1].argmax()]
    return label, freq
```

```python
test_point = np.array([9, 5])
label, freq = knn(test_point, X_train, y_train)
print("Predicted label id: ", label)

plt.scatter(dist_01[:, 0], dist_01[:, 1], marker = '.', label = 'Class 0')
plt.scatter(dist_02[:, 0], dist_02[:, 1], marker = '.', label = 'Class 1')
plt.scatter(dist_03[:, 0], dist_03[:, 1], marker = '.', label = 'Class 2')
plt.scatter(test_point[0], test_point[1], color = 'red', marker = 'x', label = 'Test Point')
plt.legend()
plt.show()
```

```python
# computing accuracy score for X_test
def accuracy_score(X_test, y_test, X_train, y_train):
    cnt = 0
    for i in range(X_test.shape[0]):
        label, freq = knn(X_test[i], X_train, y_train)
        if label == y_test[i]:
            cnt += 1
    return (cnt / X_test.shape[0])
```

```python
print("Accuracy score = ", accuracy_score(X_test, y_test, X_train, y_train))
```

# K-Nearest Neighbours Using sklearn

```python
# importing dependencies, fitting the model, predicting output and calculating the accuracy of predictions
from sklearn.neighbors import KNeighborsClassifier
from sklearn.metrics import accuracy_score

knc = KNeighborsClassifier()
knc.fit(X_train, y_train)
pred = knc.predict(X_test)

print("Accuracy score = ", accuracy_score(y_test, pred))
```
