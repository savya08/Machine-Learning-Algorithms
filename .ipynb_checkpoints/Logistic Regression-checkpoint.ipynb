# Logistic Regression

```python
import numpy as np
import matplotlib.pyplot as plt
```

```python
# creating data
mean_01 = np.array([0.0, 1.0])
cov_01 = np.array([[1.5, 0.1], [0.1, 1.4]])
dist_01 = np.random.multivariate_normal(mean_01, cov_01, 500)
dist_01 = np.hstack((dist_01, np.zeros((dist_01.shape[0], 1))))

mean_02 = np.array([5.0, 6.0])
cov_02 = np.array([[1.5, 0.1], [0.1, 1.4]])
dist_02 = np.random.multivariate_normal(mean_02, cov_02, 500)
dist_02 = np.hstack((dist_02, np.ones((dist_01.shape[0], 1))))

data = np.vstack((dist_01, dist_02))
data = np.hstack((np.ones((data.shape[0], 1)), data))
np.random.shuffle(data)
```

```python
# train-test-split
split_factor = 0.85
split = int(split_factor*data.shape[0])

X_train = data[:split, :-1]
y_train = data[:split, -1].reshape((-1, 1)).astype(np.int)
X_test = data[split: , :-1]
y_test = data[split: , -1].reshape((-1, 1)).astype(np.int)
```

```python
# visualising data
plt.scatter(dist_01[:, 0], dist_01[:, 1], marker = '.', label = 'Class 0')
plt.scatter(dist_02[:, 0], dist_02[:, 1], marker = '.', label = 'Class 1')
plt.legend()
plt.show()
```

```python
# logistic regression
def sigmoid(z):
    return 1 / (1 + np.exp(-1*z))

def hypothesis(X, theta):
    return sigmoid(np.dot(X, theta))

def error(X, y, theta):
    h = hypothesis(X, theta)
    l = np.sum(y*np.log(h) + (1 - y)*np.log(1 - h))
    return (-1*l)

def gradient(X, y, theta):
    h = hypothesis(X, theta)
    grad = np.dot(X.transpose(), (y - h))
    return grad

def gradientAscent(X, y, u = 0.002):
    theta = np.random.random((X.shape[1], 1))
    error_list = []
    max_itr = 2500
    for itr in range(max_itr):
        theta = theta + u*gradient(X, y, theta)
        error_list.append(error(X, y, theta))
    return theta, error_list
```

```python
theta, error_list = gradientAscent(X_train, y_train)
print("Bias = ", theta[0])
print("Coefficients = ", theta[1:].reshape((1, -1)))
```

```python
# visualising change in error
plt.plot(error_list)
plt.show()
```

```python
# visualising decision boundary
x1 = X_train[:, 1]
x2 = -1*((theta[0] + theta[1]*x1) / theta[2])

plt.scatter(dist_01[:, 0], dist_01[:, 1], marker = '.', label = 'Class 0')
plt.scatter(dist_02[:, 0], dist_02[:, 1], marker = '.', label = 'Class 1')
plt.legend()
plt.plot(x1, x2, color = 'gray')
plt.show()
```

```python
# predicting output for X_test and computing accuracy of prediction
def predict(X, theta):
    confidence = hypothesis(X, theta)
    y_pred = np.round(confidence).astype(np.int)
    return y_pred
y_pred = predict(X_test, theta)

accuracy_score = np.sum(y_pred == y_test) / y_test.shape[0]
print("Accuracy score = ", accuracy_score)
```

# Logistic Regression Using sklearn

```python
# importing dependencies, fitting the model, predicting output and calculating the accuracy of predictions
from sklearn.linear_model import LogisticRegression
from sklearn.metrics import accuracy_score

lr = LogisticRegression()
lr.fit(X_train[:, 1:], y_train.reshape((y_train.shape[0], )))
pred = lr.predict(X_test[:, 1:])

b = lr.intercept_
w = lr.coef_
print("Bias = ", b)
print("Coefficients = ", w)

x1 = X_train[:, 1]
x2 = -1*((b[0] + w[0][0]*x1) / w[0][1])
plt.scatter(dist_01[:, 0], dist_01[:, 1], marker = '.', label = 'Class 0')
plt.scatter(dist_02[:, 0], dist_02[:, 1], marker = '.', label = 'Class 1')
plt.legend()
plt.plot(x1, x2, color = 'gray')
plt.show()

score = accuracy_score(y_test, pred)
print("Accuracy score = ", score)
```
